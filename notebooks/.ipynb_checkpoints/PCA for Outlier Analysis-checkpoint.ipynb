## PCA for outlier analysis

To follow up my outlier analysis, I will be conducting a PCA with neutral loci only, with all outliers, and with outliers detected by two or more programs (I have run outlier analysis using Bayescan, OutFLANK, and PCAdapt). 

This requires:

[

1. Subsetting genepop files to include only the following loci:
   - neutral loci
   - all outliers 
   - outliers detected in two or more programs
2. Running PCA on all three files

] 

x 3 (for East, West, West v. East)


<br>


### Subset genepop files

With GenepopEdit, you could do this in R. However, my genotypes are encoded from 1-4, not just 1 and 2, and so GenepopEdit cannot handle my data. *If you are using GenepopEdit, see notes below*. 


Instead, I'll use the two versions of Dan's subsetting script. 

- this script is interactive and must be run from the command line
- in the output file, the following need to be fixed: (1) add the first "Pop" line after loci list. (2) remove extra space at the end of each row of genotypes.


[subsetGenepop.py]() takes a list of loci to keep (original)
<br>
[subsetGenepop.py]() takes a list of loci to remove (altered for this analysis; I checked results in excel after first run.)

#### Notes on GenepopEdit

[SubsetGenepop_Genepopedit.R]()

GenepopEdit requires all population names to be formatted as `POP01_`, with all individual sample IDs in the same population starting with the same name. Currently, my western individuals have individual IDs coded by sampling site + sampling date, which means that within a single "population" I have multiple population IDs. To fix this, I gave all samples at the same site the same population ID, and then coded the individual IDs to indicate which date they belonged to -- 
<br>
first collection date = no individual ID # change [ 01 = 01 ]
<br>
second collection date = `1` + individual ID # [ 01 = 101 ]
<br>
third collection date = `2` + individual ID # [ 01 = 201 ] 
<br>
etc...

<br>
Other fun things about GenepopEdit input file formats:
1. The sample ID has to be separated from the first genotyped by : space-comma-space-space
2. New lines must be designated by \n, not \r\n
