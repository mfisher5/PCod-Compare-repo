## Fis values - stacks batch 4

For this batch, I do not want to use Marine's script to correct genotypes. To see if (and how badly) stacks is undercalling heterozygotes, I need to check Fis values. 
<br>
<br>

#### 10/26/2017

### Fis per population (GENEPOP) - Uncorrected

I ran genepop `option 5 >> suboption 2`

Output file type: `.DIV`



```python
pwd
```

```python
cd ../stacks_b4_wgenome
```

```python
infile = open("batch_4_filteredMAF_filteredLoci50_filteredIndivids50_filteredHWE_noreps_genepop.txt.DIV", "r")

line = infile.readline()
while not line.startswith("Statistics per sample over all loci with at least two individuals typed:"):
    line = infile.readline()

infile.readline()
infile.readline()
print infile.readline()

pops = []
fis = []
i = 1
for line in infile:
    if i < 18:
        linelist = line.strip().split()
        pops.append(linelist[0].split("_")[0])
        fis.append(float(linelist[3]))
    i += 1
infile.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpos = list(range(len(pops)))
plt.bar(xpos, fis)
plt.grid
plt.ylabel("Fis")
plt.xticks(xpos, pops, rotation = 90)
plt.show()
```

```python
plt.close()
```
