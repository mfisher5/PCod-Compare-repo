#### Notes on GenepopEdit

[SubsetGenepop_Genepopedit.R]()

GenepopEdit requires all population names to be formatted as `POP01_`, with all individual sample IDs in the same population starting with the same name. Currently, my western individuals have individual IDs coded by sampling site + sampling date, which means that within a single "population" I have multiple population IDs. To fix this, I gave all samples at the same site the same population ID, and then coded the individual IDs to indicate which date they belonged to -- 
<br>
first collection date = no individual ID # change [ 01 = 01 ]
<br>
second collection date = `1` + individual ID # [ 01 = 101 ]
<br>
third collection date = `2` + individual ID # [ 01 = 201 ] 
<br>
etc...

<br>
Other fun things about GenepopEdit input file formats:
1. The sample ID has to be separated from the first genotyped by : space-comma-space-space
2. New lines must be designated by \n, not \r\n
